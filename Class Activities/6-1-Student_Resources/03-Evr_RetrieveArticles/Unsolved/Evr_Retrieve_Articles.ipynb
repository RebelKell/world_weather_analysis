```python
# Dependencies
import requests
from confignyt import api_key

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Store a search term
search_filters = "Portland" + "&" + "pub_date=20210101" + "&" + "end_date=20210105" 

# Build url
search_url = url + "&api-key=" + api_key + "&q=" + search_filters 
#queryULR using fstring: f"{url}?api-key={api_key}&q={search_filters}
print(search_url)
```

```python
# Retrieve articles
articles = requests.get(search_url).json()
#turn these into a list
articles_list = [article for article in articles["response"]["docs"]]
print(articles_list)
```

```python
print("Your reading list")
for article in articles_list:
    print(article["web_url"])
```

```python
for article in articles_list:
    print(f"A snippet from: {article["snippet"]}")
    print('------------')
```
